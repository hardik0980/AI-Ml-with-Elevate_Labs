# Preprocessing : 

# 1 . Handle null values 

```python
house_dataset.isna().sum()
```

# 2. Handle categorical cloumns

```python
categorical_col
```

```python
from sklearn.preprocessing import LabelEncoder

for i in categorical_col:
    lb = LabelEncoder()
    house_dataset[i] = lb.fit_transform(house_dataset[i])

categorical_col_update = house_dataset.select_dtypes(include=["O"]).columns
categorical_col_update
```

# 3. detect Outlines and Remove

```python
for i in house_dataset:
        Q1 = house_dataset[i].quantile(0.25)
        Q3 = house_dataset[i].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        house_dataset = house_dataset[(house_dataset[i] >= lower) & (house_dataset[i] <= upper)]
```

# 4. Features Scaling 

```python
numerical = numerical_col.drop("price")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
house_dataset[numerical] = scaler.fit_transform(house_dataset[numerical])
house_dataset
```

# 5. Split into features and target col

```python
x = house_dataset.drop(columns=["price"] , axis=1)
y = house_dataset.price
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train ,y_test = train_test_split(x,y,test_size=0.2 ,random_state=10)
```

# 6. import and fit Regression model & the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
print(" Predict the model :\n",y_pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("\n Evaluation the model : \n")
mse = mean_squared_error(y_test,y_pred)
print("mean square error is : \n",mse)

mae = mean_squared_error(y_test,y_pred)
print("\n mean absolute error is : \n",mae)

r2_score = r2_score(y_test,y_pred)
print("\n R2 score is : \n",r2_score)
```

# Plot regression line and interpret coefficients

```python
coefficients = pd.DataFrame({
    "Feature": x_train.columns,
    "Coefficient": model.coef_,
    "intercept":model.intercept_,
  
})
coefficients
```

```python
import warnings
warnings.simplefilter("ignore")


plt.figure(figsize=(10, 6))
sns.barplot(data=coefficients, x="Coefficient", y="Feature", palette="coolwarm")
plt.title("Linear Regression Coefficients")
plt.xlabel("Coefficient Value")
plt.ylabel("Feature")
plt.axvline(0, color="gray", linestyle="--")
plt.tight_layout()
plt.show()
```
